```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def ho(concentrations):
    ck = concentrations[0]
    cu = concentrations[1]
    cth = concentrations[2]
    
    ruk = ck / cu
    ruth = cth / cu
    hk = 3.48e-9
    hu = 9.81e-5
    hth = 2.64e-5
    return cu * (hu + ruk*hk + ruth*hth)

verena_8 = [4.7e-3, 6e-7, 3.65e-6]
verena_9 = [3e-3, 4.6e-7, 7.6e-7]
verena_10 = [4e-2, 2.2e-6, 6.5e-6]
ho_v8 = ho(verena_8)
ho_v9 = ho(verena_9)
ho_v10 = ho(verena_10)
ho_v8, ho_v9, ho_v10
```
